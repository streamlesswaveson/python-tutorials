```python
import folium
```

```python
# Folium uses open street map tiles as the default, unless 'tiles' param 
# included. 
my_map = folium.Map(location=[45.372, -121.697], zoom_start=10, tiles='Stamen Terrain')
```

```python
my_map
```

```python
folium.Marker(location=[45.3288, -121.6625],
              popup='Mt. Hood Meadows',
             icon=folium.Icon(color='green')).add_to(my_map)
folium.Marker(location=[45.3311, -121.7311],
              popup='Timberlake Lodge',
             icon=folium.Icon(color='blue')).add_to(my_map)

my_map
```

```python
import pandas
```

```python
df = pandas.read_csv('./data/Volcanoes-USA.txt')
df
```

```python
for lat,lon,name,elev in zip(df['LAT'], df['LON'], df['NAME'],df['ELEV']):
    
    marker_color = 'orange' if elev in range(0,1000) else 'red' if elev in range(1000,3000) else 'green'
    folium.Marker(location=[lat, lon],
              popup=name,
             icon=folium.Icon(color=marker_color)).add_to(my_map)
```
