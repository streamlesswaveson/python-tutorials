```python
import pandas
import folium
df = pandas.read_csv('./data/Volcanoes-USA.txt')
df
```

```python
my_map = folium.Map(location=[df['LAT'].mean(), df['LON'].mean()], zoom_start=10, tiles='Stamen Terrain')

def make_color(elev):
    if elev in range(0,1000):
        return 'green'
    if elev in range(1000,3000):
        return 'blue'
    return 'red'

for lat,lon,name,elev in zip(df['LAT'], df['LON'], df['NAME'],df['ELEV']):
    
    marker_color = make_color(elev)
    folium.Marker(location=[lat, lon],
              popup=name,
             icon=folium.Icon(color=marker_color)).add_to(my_map)
```
